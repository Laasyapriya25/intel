```python
!pip install googletrans==4.0.0-rc1
import random
from googletrans import Translator

# Initialize Google Translate
translator = Translator()

# Languages for translation
languages = {
    "hi": "Hindi",
    "kn": "Kannada",
    "ta": "Tamil",
    "te": "Telugu"
}

# Generate random vitals
def get_vitals():
    return {
        "temperature": round(random.uniform(97.5, 99.5), 1),
        "bp_systolic": random.randint(110, 140),
        "bp_diastolic": random.randint(70, 90),
        "spo2": random.randint(90, 100),
        "pulse": random.randint(60, 100)
    }

# Simple AI diagnosis logic
def ai_diagnosis(vitals):
    if vitals["spo2"] < 94:
        return "Possible hypoxia detected. Suggest oxygen therapy."
    elif vitals["bp_systolic"] > 130:
        return "High BP - Suggest physician consult for hypertension."
    else:
        return "Vitals normal. No urgent condition detected."

# Get vitals and diagnosis
vitals = get_vitals()
diagnosis_en = ai_diagnosis(vitals)

# Print Vitals and Diagnosis in English
print("\nü©∫ Vitals:", vitals)
print("üí° Diagnosis (English):", diagnosis_en)

# Translate diagnosis into multiple languages
print("\nüåê Translated Diagnosis:")
for lang_code, lang_name in languages.items():
    try:
        translated = translator.translate(diagnosis_en, dest=lang_code).text
        print(f"{lang_name}: {translated}")
    except Exception as e:
        print(f"{lang_name}: [Translation Error - {e}]")

```
